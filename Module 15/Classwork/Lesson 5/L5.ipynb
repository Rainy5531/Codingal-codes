```python
import pandas as pd
import seaborn as sns
```

```python
weather = pd.read_csv('Weather Dataset.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
weather.isnull().sum()
```

```python
weather.dropna(axis=1, how='all', inplace=True)
```

```python
weather.isnull().sum()
```

```python
sns.barplot(x="weather_type", y="temperature", data=weather)
```

```python
sns.histplot(weather['temperature'], kde=True)
```

```python
sns.displot(weather['humidity'], rug=True)
```

```python
sns.jointplot(x="humidity", y="temperature", data=weather)
```

```python
sns.jointplot(x="humidity", y="temperature", data=weather, kind="hex")
```

```python
sns.jointplot(x="humidity", y="temperature", data=weather, kind="kde")
```

```python
sns.pairplot(data=weather, vars=["humidity", "temperature", "air_pollution_index"])
```

```python
sns.stripplot(x="weather_type", y="temperature", data=weather)
```

```python
sns.boxplot(x=weather['humidity'], y = weather['temperature'], hue=weather['weather_type'])
```

```python
sns.countplot(x="weather_type", data=weather)
```

```python
sns.lmplot(x="humidity", y='temperature', data=weather)
```

```python
sns.lmplot(x="humidity", y='temperature', data=weather, hue='weather_type')
```
